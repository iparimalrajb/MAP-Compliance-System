```python
import pandas as pd
```

```python
df=pd.read_json("PL_Table.json",lines=True)
```

```python
df.head()
```

```python
def extracting_category_subcategory(pl_value):
    parts = pl_value.split("_")

    if len(parts) >= 3:
        category = parts[1]
        subcategory = parts[2]
    else:
        category = None
        subcategory = None

    return pd.Series([category, subcategory])
```

```python
df[["category","subcategory"]] = df["PL"].apply(extracting_category_subcategory)
```
