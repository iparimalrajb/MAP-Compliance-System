```python
import pandas as pd
```

```python
df_base = pd.read_csv("Base_Price_monitoring_table.csv")
df_seller = pd.read_csv("table1.csv")
df_seller_map = pd.read_csv("Sellers_Mapping_Table_Cleaned.csv")
```

```python
df_seller.rename(columns={"SKU": "sku"}, inplace=True)
```

```python
df_monitor = df_base.merge(df_seller[["sku", "Marketplace", "Seller_name", "region", "adv_price"]],on="sku",how="left")
```

```python
df_monitor = df_monitor.merge(df_seller_map,on="Seller_name",how="left")
```

```python
exchange_rates = {
    "USA": 91,
    "UAE": 25,
    "UK": 122,
    "IND": 1,
    "CAN": 66
}
```

```python
df_monitor["Advertised_price_INR"] = (df_monitor["adv_price"] *df_monitor["region"].map(exchange_rates))
```

```python
df_monitor.head(100)
```

```python
df_monitor["Violation date"] = pd.to_datetime("21/12/2025", dayfirst=True)
```

```python
df_export = df_monitor[[
    "sku",
    "Violation date",
    "PL",
    "category",
    "subcategory",
    "MAP",
    "LPP",
    "season",
    "promotion",
    "Marketplace",
    "Seller_name",
    "Homologated_sellers_name",
    "region",
    "adv_price",
    "Advertised_price_INR"
]]
```

```python
df_export.to_csv("Price_monitoring_table1.csv", index=False)
```
